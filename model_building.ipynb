# Build a regression model.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
analytics_df = pd.read_csv("analytics_df.csv")
```

__Build simple linear regression model for each of the column in analytics dataframe and the number of bikes__

```python
y = pd.Series(analytics_df["number of bikes"])
for column in analytics_df[['avg distance','avg review count','avg rating','avg price','restaurant count','high rating count']]:
    print("\033[1m" + "Simple Linear Regression Model between " + column + " and number of bikes" + '\033[0m')
    X = analytics_df[column]
    X = sm.add_constant(X)
    lin_reg = sm.OLS(y,X)
    model = lin_reg.fit()
    print_model = model.summary()
    print(print_model)
    print(" ")
    print(" ")
```

> _these models have significantly low R squared, low coef and super high p value, so we can reject these model_

__Building all possible multivatiate linear regression model__

```python
from itertools import combinations
variables = ['avg distance','avg review count','avg rating','avg price','restaurant count','high rating count']
for numberOfVariable in [2,3,4,5,6]:
    for columns in combinations(variables, numberOfVariable):
        print("\033[1m" + "Simple Linear Regression Model between ", columns , " and number of bikes" + '\033[0m')
        X = analytics_df[list(columns)]
        X = sm.add_constant(X)
        lin_reg = sm.OLS(y,X)
        model = lin_reg.fit()
        print_model = model.summary()
        print(print_model)
        print(" ")
    
```

>_These models have low R square and high p value, so we can reject these models_
